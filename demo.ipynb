```python
import DidacticFramework as df

# Create a Data Loader object that will feed the optimizer data
loader = df.DataLoader()
loader.generate_dataset(func="Discontinuity") # Create data with a built-in dataset

# Create a Neural Network object to store the model
network = df.Network(layer_sizes=[1, 64, 64, 64, 1], activation_function='relu')

# Create an Optimizer object to facilitate training
optimizer = df.Optimizer(learning_rate=0.01, num_epochs=500, batch_size=512, drop_prob=0.2, drop_decay_rate=2)

print("Initial Parameter Values:")
network.visualize_network()

# Train the network
_, _ = optimizer.train_network(network, loader, grad_checking=False)
```

```python
# Characterize the trained network using a separate "test" dataset:
optimizer.test_network(network, loader)

print("Trained Network Characterization (on Test Set):")
network.visualize_network()
```
